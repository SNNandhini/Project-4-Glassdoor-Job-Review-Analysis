```python
#importing library
import pandas as pd
from sqlalchemy import create_engine
from config import user
from config import password2
from urllib import parse
```

```python
# Creating a reference to the CSV files
data_csv = "../Resources/glassdoor_reviews_clean.csv"
data_csv = pd.read_csv(data_csv)
data_csv
```

```python
# Remoing all rows with null values
data_csv=data_csv.dropna()
data_csv
```

```python
data_csv = data_csv.drop("Unnamed: 0", axis=1)
```

```python
data_csv
```

```python
# Checking for duplicates
data_csv.duplicated().sum()
```

```python
# Removing the duplicates
data_csv = data_csv.drop_duplicates()
data_csv
```

```python
# Checking for the data types
data_csv.info()
```

```python
# Checking for  number of  unique values across all columns
data_csv.nunique()
```

```python
# Checking for the max length of data (used while creating table in SQL)
data_csv.cons.str.len().max()
```

```python
data_csv.to_csv("../Resources/glassdoor_cleaned_data.csv")
```

# Create Database Connection

```python
# Creating a connection string for the Glassdoor database
protocol="postgresql"
username=user
password=parse.quote(password2)
host="localhost"
port=5432
database_name="Glassdoor_Reviews"
connection_string=f"{protocol}://{username}:{password}@{host}:{port}/{database_name}"
```

```python
# Creating an Engine instance
engine=create_engine(connection_string)
engine.table_names()
```

```python
# Uploading data into SQL from the dataframe
data_csv.to_sql(name="glassdoor",con=engine,if_exists="append", index=False)
```
